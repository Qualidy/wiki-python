## Variablen

__Variablen__ sind ein wichtiger Baustein beim Programmieren. 

Variablen geben Dinge Namen:

```python
text = "Hallo Welt"
```

Die beiden Wörter "Hallo Welt" bekommen den Namen `text`. Damit kann man sie im weiteren Programmverlauf wieder verwenden:

```python
print(text)
```

Hinweise: 
- _Variablen sollten aussagekräftige Namen haben_ (Lange Variablennamen kosten kein Geld ;))
- Variablennamen dürfen keine Leerzeichen haben

# Ausgabe von Werten

Natürlich kann man Variablen auch auf der Kommandozeile ausgeben. Dafür nutzt man die `print` Funktion, die ihr bereits kurz kennen gelernt habt:

````python
print("Hallo Welt")

text = "Hallo Welt"
print(text)

dein_name = "<NAME EINTRAGEN>"
print(f"Hallo {dein_name}")    
````

In der letzten Zeile seht ihr wie man Daten aus Variablen ausgeben kann. Mit einem `f` vor der Zeichenkette sagt ihr Python, dass in dieser Zeichenkette (einem __f-string__) Variablen genutzt werden sollen. 

Variablen werden mit Hilfe geschweifter Klammern (`{}`, engl. _curly braces_) in einen f-String eingebettet. Beim Ausführen des Programms werden diese dort einfach ersetzt.

# Eingaben von der Kommandozeile

Programme, die keine Eingaben von außen ermöglichen, können nicht wirklich viel machen. Eingaben können auf viele verschiedene Varianten passieren:

- Texteingabe auf der Kommandozeile
- Einlesen von Dateien, die die Eingabe enthalten
- Eingaben übers Netzwerk

Hier konzentrieren wir uns auf die erste Option: Eingaben, die auf der Kommandozeile passieren.

Der Python-Befehl dafür ist `input`:

```python
eingabe = input()
print("Eingabe:")
print(eingabe)
```

Das Programm fragt zu erst nach einer Eingabe durch den Benutzer und speichert diese Eingabe in der Variable `eingabe`. Im nächsten Schritt wird eure Eingabe einfach wieder ausgegeben. Probiert es selbst aus:

````python
eingabe = input()
print("Eingabe:")
print(eingabe)
````
