Vereinfach kann man sagen, dass alle Grundrechenarten genauso funktionieren, wie man es erwartet. Hier noch mal als Übersicht:

Mathematische Operation | Symbol | Beispiel
------------------------|--------|----------
Addition                |   +    | `5 + 6`, `6.0 + 7.0`
Subtraktion             |   -    | `6 - 10`

Division                |   /    | `8 / 6 == 1.5`
ganzzahlige Division    |   //   | `8 // 6 == 1` 
Modulo (Restberechnung) |    %   | `10 % 3 == 1`
Potenzieren             |   **   | `5**2 == 25`

Zusätzlich gibt es Vergleichsoperatoren:

Vergleich     | Symbol     | Beispiel
--------------|------------|----------
Kleiner       |  <         |  `5 < 10`
Kleiner gleich|  <=        |  `5 <= 10`
Größer        |  >         |  `10 > 0`
Größer gleich |  >=        |  `10 >= 10`
Gleich        |  ==        |  `10 == 10`
Ungleich      |  !=        |  `1 != 10`

# Zuweisungen

Wenn eine Variable einen neuen Wert bekommt, nennt man das __Zuweisung__. Die Syntax dafür kennen wir schon: `=`.

Beispiele:

```python
# Beispiel 1
x = 5     # x wird der Wert 5 zugewiesen. Das kennen wir von gestern
x = x + 5 # x wird um 5 erhöht. Als erstes wird die rechte Seite ausgeführt und dann der linken Seite zugewiesen. 
x += 5    # x wird ebenfalls um 5 erhöht. Kurzschreibweise.

import random       # importiere das Modul random (später mehr dazu)
x = random.random() # x kann auch der Wert eines Funktionsaufrufs zugewiesen werden. 

# Beispiel 2
a = 10
b = 1
c = a + b
```

# Datentypen

Jede Variable hat einen __Datentyp__. Aus Sicht von Python, sind ganze Zahlen (`integer`) etwas anderes als Zeichenketten (`str`) und (Gleit-)Kommazahlen (`float`).

Jeder Datentyp hat einen bestimmten Verwendungszweck:

**Primitive Datentypen** sind in Python direkt eingebaut und fester Bestandteil der Sprache. Zusätzlich gibt es auch noch **zusammengesetzte Datentypen**, die mehrere Informationen in einer Einheit zusammenfassen. Hier eine Auswahl der Datentypen in Python, die immer vorhanden sind:

Deutsche Bezeichung | Fachbegriff | Python  | Art             |Beispiel
--------------------|-------------|---------|-----------------|-----------
Ganze Zahlen        | Integer     | `int`   | prmitiv         |`1`, `5329`
Fließkommazahlen    | Float       | `float` | prmitiv         | `1.2`
Zeichenketten       | String      | `str`   | zusammengesetzt | `"Hallo Welt"`
Wahrheitswerte      | Boolean     | `bool`  | prmitiv         |`True`, `False`
Listen              | List        | `list`  | zusammengesetzt |`[1,2,3,4]`

Beispiele:

```python
is_boolean = True
a_string = "Hallo Wolfsburg"
floatyMcFloatface = 3.14
number = 1
```

Die Datentypen werden in Python typischerweise nicht explizit angegeben. Seit einigen Jahren können diese auch explizit definiert werden:

```python
a: str = "Hallo Wolfsburg"
b: int = 12
```

## Datentyp einer Nutzereingabe

Mit `input` können wir Nutzereingaben einlesen (manche Tools haben Schwierigkeiten mit der `input` Funktion in  Jupyter Notebooks):

```bash

print(z)
```

```bash

Eingabe: 20

```

**Frage**: Was erwartest du als Datentyp von `z`?

Obwohl wir einen Integer eingelesen haben, wird in `z` zunächst ein String gespeichert. Von der Kommandozeile werden *immer* Strings eingelesen. Erwartet man eine Zahl, muss man diese zunächst von einem String in einen Integer oder eine Fließkommazahl umwandeln. 

Wie das geht siehst du hier: 

```python
text = "12"
ganzzahl = int(text)    # == 10
kommazahl = float(text) # == 10.0
```
