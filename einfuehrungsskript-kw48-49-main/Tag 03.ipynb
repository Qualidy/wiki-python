## Besondere Bedingungen

In Python gibt es noch einige weitere praktische Wege mit denen man Bedingungen schreiben kann.

Zum Beispiel kannst du einfach überprüfen, ob ein Wert in einer Liste ist. Im Beispiel wird überprüft, ob die Zahl `20` in der Liste vorhanden ist oder nicht.

```python
liste = [10, 20, 30, 40, 50]
if 20 in liste:
    print("20 gefunden!")
```

Das ganze ist auch mit Zeichenketten möglich:

```python
text = "Das ist ein Test!"

# Versuch 1
if "e" in text:
    print("e gefunden!")

# Versuch 2
if "Das" in text:
    print("Das gefunden!")
```

Bei Zeichenketten kann man sogar nach dem Vorhandensein mehrere Buchstaben gleichzeitig suchen, die dann auch genauso im Text vorhanden sein müssen, damit die Bedingung `True` zurückgibt.

````python
liste = [1, 2,3,4,5,6,7,9,10]
x = 1
````

````python
if x in liste and 5 in liste:
    print("ja")
````

````python
if 1 in liste or 5 in liste:
    print("ja")
````

# Zufallszahlen

Um die heute Tagesaufgabe lösen zu können, benötigst du noch eine Sache: Zufallszahlen! Hier gibts kurz ein paar Infos, wie du mit Python Zufallszahlen erzeugen kannst.

In Python gibt es ein Modul `random` welches man importieren kann:

```python
import random 
```

Damit kann man ganz allgemein zusätzliche Funktionalitäten importieren, die man dann einfach nutzen kann. Im Modul `random` gibt es zum Beispiel die Funktion `random()` und `randint(lower, upper)`.

- `random()`: Erzeugt eine Fließkomma-Zufallszahl zwischen 0 und 1.
- `randint(lower, upper)`: Erzeugt eine ganzzahlige Zufallszahl zwischen `lower` und `upper`, wobei `upper` eingeschlossen ist.

Möchtest du also eine ganzzahliger Zufallszahl zwischen 0 und einschließlich 100, kannst du das wie folgt machen:

```python
zufallszahl = random.randint(0, 100)
```

Bevor man die Funktion nutzen kann, muss man den Modulnamen mit angeben. 

`random.randint(0, 100)` sagt damit, dass die Funktion `randint` im modul `random` ist. Wieso ist das nötig? Weiß Python nicht, dass das dort drin ist? 

Nein, das weiß Python nicht. Stelle dir vor, dass du ein weiteres Modul importierst, welches auch eine Funktion beinhaltet, die `random` heißt. Wie soll Python zwischen diesen beiden Funktionen unterscheiden, wenn kein Modul angegeben ist?
