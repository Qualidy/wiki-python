# Verschachtelte If-Abfragen

Man kann if-Abfragen natürlich auch beliebig verschachteln:

```python
if BEDINGUNG_1:
    if BEDINGUNG_2: 
        if BEDINGUNG_3:
           ...
        else:
           ...
    else:
       ...
else:
    ...
```

Wie man allerdings auch gut sieht, ist das bereits bei 3 Ebenen schon alles andere als übersichtlich. Editoren wie VSCode oder IDEs wie PyCharm helfen dabei, dass man den Überblick nicht verliert. 

````python
n = 60
if n > 50:
    print("Hallo")
    if n < 70:
        print("blabla")
    else:
        print("ist größer als 70")
else:
    print("kleiner als 50")
````

````python
wochentage = ["Montag", "Dienstag", "Mittwoch", "Donnerstag", "Freitag", "Samstag", "Sonntag"]

tag = input("Welcher Wochentag ist heute")

if tag in wochentage:
    index = wochentage.index(tag)
    print(f"{tag} ist das {index}-Element in der Liste")
    if index == len(wochentage) - 1:  
        print("Morgen ist Montag")
    else:
        print(f"Morgen ist {wochentage[index + 1]}")

else:
    print("Das ist kein Wochentag, du Held!")
            
````
