# Funktionen

Dinge, die man immer wieder braucht, kann man in Funktionen auslagern. Python bringt jede Menge solcher Funktionen mit. 

Eine davon kennt ihr schon: `print(...)`.

Funktionen können Argumente haben, die das Verhalten der Funktion bestimmen. 

- `print()` gibt einfach eine leere Zeile aus.
- `print("Hallo")` gibt `Hallo` aus
- `print("Hallo", "Du")` gibt `Hallo Du` aus.

Sehr viele Funktionen haben auch Rückgabewert. Dies kann z.B. das Ergebnis einer erfolgreichen Berechnung sein.

Auch da kennt ihr schon eine Funktion: `eingabe = input()`.

Der Rückgabewert von `input()` ist die Eingabe des Nutzers auf der Kommandozeile, die dann in der entsprechenden Variable gespeichert wird.

## Eigene Funktionen definieren

Will man eigene Funktionen definieren, z.B. eine Funktion, die den Durchschnitt dreier Zahlen berechnet, so kann man das wie folt tun:

```python
# Funktionsdefintion:
def average(a, b, c):
    sum = a + b + c
    return sum / 3

# Funktionsaufruf
result = average(2, 2, 5) # Rückgabe 3
print("Der Durchschnitt ist:", result)
```

Codeteil | Bedeutung
-|-
`def` | Das Schlüsselwort `def` zeigt an, dass hier eine Funktion definiert wird.
`average` | Das ist der Name der Funktion, mit der wir die Funktion zu einem späteren Zeitpunkt aufrufen können.
`(a, b, c)` | Hier werden die Parameter der Funktion definiert. In diesem Fall drei (`a`, `b` und `c`). Diese müssen beim Methodenaufruf mit konkreten Werten gefüllt werden.
`:` | Der Doppelpunkt zeigt an, dass der **Funktionskopf**, der den Funktionsnamen und die Parameter enthält, nun definiert ist und in der nächsten Zeile der **Funktionsrumpf** beginnt. Alles was nach dem `:` eingerückt ist, gehört zum Funktionsrumpf.
`return` | Das Ergebnis dieser Zeile wird als Ergebnis bzw. **Rückgabewert** der Funktion beim Aufruf zurück gegeben und kann z.B. in einer Variablen gespeichert werden.

# Verarbeitung von Zeichenketten

Hier soll es einen ganz kurzen Überblick darüber geben, wie man Zeichenketten verarbeiten kann, vor allem auf welche Wege man diese Zusammenfügen kann.

## Formatierung

### Weg 1 - f-Strings

Die einfachste und modernste Methode Zeichenketten zu vereinen, sind sogenannte **f-Strings**:

```python
name = "Thomas"
greeting = f"Hallo {name}"
print(greeting)
```

Die Variable `greeting` enthält dann eine Zeichenkette "Hallo Thomas". Durch die geschweiften Klammern und das vorangestellt `f` weiß Python, dass es in der Zeichenkette den Wert der Variable `name` eintragen muss. In den geschweiften Klammern sind die meisten Python-Ausdrücke erlaubt. Du kannst also auch darin rechnen, vergleichen und viele andere Dinge machen.

f-Strings funktionieren auch mit mehrzeiligen Strings:

```python
f"""
Hallo {name},

Glückwunsch zu deinem {alter}sten Geburtstag!

Grüße {mein_name}
"""
```

### Weg 2 - der + Operator

Alternativ kann man mit `+` Zeichenketten zusammensetzen. Das geht auch, ist aber vor allem dann unbequem, wenn man viele Zeichenketten miteinander verbindet:

```python
name = "Thomas"
greeting = "Hallo " + name
```

Hier sieht man auch, dass das Leerzeichen hinter `Hallo` mit angeben werden muss. Das  macht diese Technik oft umständlicher und fehleranfälliger als andere.

Der Fachbegriff für das zusammensetzen von Strings ist übrigens *konkatinieren*.

### Weg 3 - Formatter

Die Nutzung der `format` Methode ist der Vorgänger des f-Strings. Diese gilt jedoch als umständlicher und weniger übersichtlich:

```python
# Hallo Thomas, du kommst aus Wolfsburg
text = "Hallo {1}, du kommst aus {0}".format("Wolfsburg", "Thomas")
```

Die geschweiften Klammern geben wieder zu ersetzene Teile des Strings an. Die Zahlen in den geschweiften Klammern stehen für die Positionen der Argumente an die `format`-Funktion. 

### Escape-Zeichen

Mit Escape-Zeichen `\` kann man die Ausgabeform kontrollieren. Hier nur eine kurze Übersicht:

- `\n`: Neue Zeile
- `\t`: Tab
- `"Hallo \"Du\" "` gibt `Hallo "Du"` aus.
- viele weitere, aber das sind erstmal die wichtigsten.

## Slicing
String-Slicing ist eine leistungsstarke Methode in Python, um Teile einer zusammengesetzten Datentypen (wie Strings) zu extrahieren oder zu manipulieren. 

Mit String-Slicing können wir Teile eines Strings basierend auf ihren Indexpositionen auswählen. Die grundlegende Syntax dafür lautet `string[start:end]`, wobei `start` den Index des ersten Zeichens einschließt und `end` den Index des ersten Zeichens ausschließt, das nicht mehr in den ausgewählten Teil des Strings gehört. 

```python
s = "Aloa Wolfsburg"
print(s[0:3]) # Rückgabe: "Alo"
```

Wir können auch einen optionalen dritten Parameter verwenden, um Schrittweiten anzugeben. 
```python
print(s[0:13:2]) # Rückgabe "Aaosr"
```
Begründung: Jedes dritte Zeichen von `"Hallo Wolfsburg"`, wobei das erste (also `0`-te) dazu zählt: **A**lo**a**_W**o**lf**s**bu**r**g &rarr; "Aaosr"

Mit String-Slicing können wir Zeichenketten auf einfache Weise zerlegen oder bestimmte Teile für Verarbeitung oder Anzeige zu extrahieren, was in vielen Anwendungen, insbesondere bei der Verarbeitung von Text, äußerst nützlich ist.
