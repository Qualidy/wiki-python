# Einen Zahlenbereich durchlaufen

Obwohl wir uns erst morgen mit Schleifen beschäftigen werden, gibt es hier schon mal einen kleinen Ausblick.

Ganz häufig, muss man alle Zahlen in einem bestimmten Zahlenbereich nutzen, um zum Beispiel Berechnungen durchzuführen.

Dafür gibt es in Python das `for`-Konstrukt:

```python
for i in range(0,10):
    print(i)
```

In der `range` Funktion geben wir den Start und nicht-inklusiven Endwert an. Das Beispiel gibt als alle Zahlen von 0 bis einschließlich 9 aus.
