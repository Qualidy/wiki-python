# Befehle wiederholen: Schleifen

Häfig müssen bestimmte Codeteile wiederholt werden. Man weiß meist aber vorher nicht, *wie oft* diese wiederholt werden müssen.. Deswegen hat Python auch eine Möglichkeit dies zu tun: Schleifen.

Schleifen werden ausgeführt, solange eine Bedingung erfüllt ist. 

Es gibt zwei Varianten von Schleifen in Python, die wir uns näher anschauen wollen:

## For-Schleifen

Die wahrscheinlich am häufigsten verwendete Schleife, die for-Schleife.

Grundsätzlich ist eine for-Schleife wie folgt aufgebaut:

```python
for ELEMENT in SEQUENCE:
    do_something_with_element()
```

for-Schleifen eigenen sich hervorragend, um über Zahlenbereiche, Listen oder auch Strings zu iterieren. 

Um eine fortlaufende Reihe an Zahlen zu generieren, gibt es in Python die `range(start, stop)`-Funktion:

```python
for i in range(0,10):
    print(i)
```

Diese Code gibt alle Zahlen zwischen 0 und 10 (exkl. 10) aus. 

Genauso kann man mit Listen arbeiten:

```python
liste = ["Hallo", "Welt", "wie", "geht", "es", "euch"]
for elem in liste:
    print(elem)
```

führt zu folgender Ausgabe:

```bash
Hallo
Welt
wie
geht
es
euch
```

Genauso funktioniert dies auch mit Zeichenketten. Mit einer for-Schleife kann man diese Zeichen für Zeichen durchgehen:

```python
for character in "Hallo":
    print(character)
```

gibt 

```bash
H 
a
l
l
o
```
aus.


## While-Schleifen

Die zweite Variante, die es gibt, sind while-Schleifen. Sie funktionieren etwas anders:

```python
while BEDINGUNG:
    do_something()
```

`BEDINGUNG` ist genau das was wir bereits bei if-Abfragen kennengelernt haben. Die Bedingung muss wahr sein, damit der **Schleifenkörper** (`do_something()`) ausgeführt wird. Zum Schleifenkörper gehört alles, was nach dem `:` eingerückt ist.

Auch mit einer while-Schleife kann man über einen Zahlenbereich iterieren:

```python
i = 0
while i < 10:
    print(i)
    i += 1
```

Dies führt zur exakt selben Ausgabe, wie das erste Beispiel für die for-Schleife.

**Achtung**: Wie ihr sehen könnt, müssen wir die Zählvariable `i` vorher auf Null setzen und dann auch per Hand bei jedem Durchlauf um 1 erhöhen. Ansonsten haben wir eine Endlosschleife.

while-Schleifen können auch einen `else`-Zweig haben, der ausgeführt wird, wenn `BEDINGUNG` nicht mehr `True` ist:

```python
n = 0
while n < 3:
    print(n)
    n += 1
else:
    print("ENDE")
```

Die Ausgabe hier ist wie folgt:

```bash
0
1
2
ENDE
```

# if-Abfragen in Schleifen

Natürlich können if-Abfragen und Schleifen auch beliebig kombiniert werden. Häufig nutzt man eine Schleifen, um über eine Liste oder einen Bereich zu iterieren und Elemente die eine bestimmte Bedigung erfüllen werden dann ausgegeben oder weiter verarbeitet.

Das sieht dann beispielsweise so aus:

```python
for i in [1, 2, 3, 4, 5]:
    if i % 2:
        print(i)
```

Diese Beispiel gibt folgendes aus:
```bash
2
4
```

Grundsätzlich geht dies auch mit einer while-Schleife:

```python
i = 0
while i <= 5:
    if i % 2:
        print(i)
    i += 1
```

Die Ausgabe ist identisch mit der aus dem Beispiel mit der for-Schleife.

## Vorzeitiges Abbrechen der Schleife

In vielen Fällen sucht man einfach einen Wert in einem Bereich oder ein bestimmtes Element in einer Liste. Sobald man dieses gefunden hat, kann man die Schleife eigentlich abbrechen. Dafür nutzt man das Keyword `break`:

```python
for i in range(0,10):
    if i == 5:
        break
        
    print(i)
```

Dieser Code gibt folgendes aus:

```bash
0
1
2
3
4
```

Sobald die Bedingung `i == 5` wahr wird, sorgt `break` dafür, dass wir die Schleifen verlassen. Damit sparen wir uns 5 weitere Durchläufe. Bei kompexen Problemstellungen kann man damit sehr viel Zeit und damit auch Energie sparen.

Auf der anderen Seite gibt es aber auch Fälle in denen man nicht die ganze Schleife beenden will, sondern nur den aktuellen Durchlauf. Dafür nutzt man das Keyword `continue`.

```python
for i in range(0,10):
    if i != 5:
        continue
        
    print(i)
```

Im Gegesatz dazu ist die Ausgabe hier:

```bash
5
```

Wieso? Für jeder Zahl zwischen 0 und 10 wird überprüft, ob diese Zahl ungleich 5 ist. Ist sie ungleich 5, springen wir direkt zum nächsten Durchlauf. Nur für den Fall, das die Zahl gleich 5 ist, erreichen wir die print-Funktion überhaupt.

Sehr häufig wird break in Kombination mit while-Schleifen verwendet. Wieso? Weil es sehr einach ist damit Endlosschleifen zu erzeugen, die man unter bestimmten Bedinungen beenden will.

Zum Beispiel:

```python
while True:
    eingabe = input()
    if eingabe == "C":
        break

    # irgendwas mit der Eingabe machen
```

Durch `while True:` läuft diese Schleife theoretisch endlos lange, weil die Bedingung immer wahr ist. Bei jedem Schleifen durchlauf wird der Nutzer nach einer Eingabe gefragt. Sobald der Nutzer "C" eingibt, wird die Schleife durch den Befehl `break` verlassen und das Programm kann normal weiterlaufen.
