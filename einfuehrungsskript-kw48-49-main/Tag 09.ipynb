# Verschachtelte Schleifen

Natürlich kann man nicht nur verschachtelte Ebenen an if-Abfragen haben, sondern auch verschachtelte an Schleifen.

Beispiel:

```python
for i in range(0, 11):
    for j in range(0,11):
        print(f"{i} hoch {j} = {i**j}")
```

Jede Schleife hat ihre eigenen Laufvariable, die ganz normal genutzen werden kann. Das selbe geht natürlich auch mit while-Schleifen:

```python
i = 0
while i <= 10:
    i += 1
    j = 0
    while j <= 10:
        print(f"{i} hoch {j} = {i**j}")
        j += 1
```

Wir sehen aber hier schon, dass es selbst für einfache Beispiele schnell unübersichtlich werden kann.
