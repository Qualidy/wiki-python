# Aufgabe tick tack tow

Nun wollen wir das Tick Tack Toe Spiel implementieren. Hier soll in der Konsole das Tick Tack Toe feld angezeigt werden und die beiden spieler abwechselnd dazu aufgefordert werden eine Eingabe vorzunehmen. Freie Felder sollen durch eine Zahl dargestellt werden, die angibt, was eingegeben werden muss, um in dieses Feld sein zeichen zu setzen:

```
 1 | 2 | 3
---|---|---
 4 | 5 | 6 
---|---|---
 7 | 8 | 9

Spieler A ist am Zug. Gib ein freies Feld ein:
> 5

Das ist erlaubt.

 1 | 2 | 3
---|---|---
 4 | A | 6 
---|---|---
 7 | 8 | 9

Spieler B ist am Zug. Gib ein freies Feld ein:
> 5

Das ist nicht erlaubt.

Spieler B ist am Zug. Gib ein freies Feld ein:
> 4

 1 | 2 | 3
---|---|---
 B | A | 6 
---|---|---
 7 | 8 | 9

```
usw.

Das ganze ist komplizierter, als man zunächst denkt. Versuche das Problem zunächst in Unterprobleme zu zerteilen.
