```python
# 6 Punkte
# Schreibe eine Funktion, die einen String erwartet und
# jeden Kleinbuchstaben groß und jeden Großbuchstaben klein schreibt.
#
# "Aloh" -> "aLOH"
# "" -> ""
# "aaaa" -> "AAAA"
# "Ich bin ein Text" -> "iCH BIN EIN tEXT"
def swap_case(words):
    großbuchstaben1= "ABCDEFGHIJKLMNOPQRSTUVWXYZÄÖÜ"
    kleinbuchstaben1= "abcdefghijklmnopqrstuvwxyzäöü"
    new_string= ""
    for i in words:
        if i in großbuchstaben1:
            new_string+=i.lower()
        elif i in kleinbuchstaben1:
            new_string+=i.upper()
        else:
            new_string+=" "
    return new_string

swap_case("Ich bin ein Text")


```
